```python
#4.fill missing value using Group logic
df['Product_Search_Method'] =(
    df.groupby('Purchase_Categories')['Product_Search_Method']
    .transform(lambda x:x.fillna(x.mode().iloc[0])
               if not x.mode().empty else x)
)
```

```python
#5.Clean column names
df.columns = df.columns.str.strip()
df.columns = df. columns.str.replace(" ","_")
df.columns = df. columns.str.replace("__+","_",regex =True)
df.columns
```

```python
#6.Rename specific columns
df= df.rename(columns={
    'age':'Age',
    'Add_to_Cart_Browsing':'Add_To_Cart_Browsing' 
})
```

```python
df.columns
```

```python
#7.Handles duplicate columns name
cols=df.columns.tolist()
indexes = [
    i for i , c in enumerate(cols)
    if c == 'Personalized_Recommendation_Frequency'
]
if len(indexes)>1:
    cols[indexes[1]]='Personalized_Recommendation_Frequency_Alt'
df.columns =cols
df.columns
```

```python
#8.Timestamp cleaning
df['Timestamp'] = pd.to_datetime(
    df['Timestamp'],
    format='%Y/%m/%d %I:%M:%S %p GMT+5:30',
    errors='coerce'
)


```

```python
df['Timestamp'] = df['Timestamp'].dt.tz_localize(None)

```

```python
df['Date'] = df['Timestamp'].dt.date
df['Day_Name'] = df['Timestamp'].dt.day_name()
df['Time'] = df['Timestamp'].dt.strftime('%I:%M:%S %p')
```

```python
df[['Timestamp', 'Date', 'Day_Name', 'Time']].head()
```

```python
first_cols = ['Date', 'Day_Name', 'Time']
remaining_cols = [col for col in df.columns if col not in first_cols]

df = df[first_cols + remaining_cols]
```

```python
df.head()
```

```python
df.columns
```

```python
#9.Standardize all Text data
exclude_cols = ['Date', 'Time','Age','Personalized_Recommendation_Frequency_Alt','Rating_Accuracy','Shopping_Satisfaction']   

cat_cols = [
    col for col in df.select_dtypes(include='object').columns
    if col not in exclude_cols
]

for col in cat_cols:
    df[col] = df[col].astype(str).str.strip().str.lower()
```

```python
#10.Handle remaining missing values
#categorical = mode
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0],inplace=True)
```

```python
#Numerical= median
num_cols=df.select_dtypes(include=['int64','float64']).columns
for col in num_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median,inplace=True)
```

```python
df.isnull().sum()
```

```python

# 11.Age cleaning and Age_Group creation

# convert age to numeric
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# filter valid age range and make explicit copy
df = df[(df['Age'] >= 10) & (df['Age'] <= 80)].copy()

# create age groups
labels = ['Young_Adult', 'Adult', 'Middle_Aged', 'Senior']
df['Age_Group'] = pd.qcut(df['Age'], q=4, labels=labels)

# adjust column position
cols = list(df.columns)
age_index = cols.index('Age')
cols.insert(age_index + 1, cols.pop(cols.index('Age_Group')))
df = df[cols]
```

```python
df[['Age', 'Age_Group']].head()
df['Age_Group'].value_counts()
```

```python
df[['Age','Age_Group']].head()
```

```python
#12.Standardize frequency columns

df['Purchase_Frequency'] = df['Purchase_Frequency'].replace({
    'few times a month': 'few_month',
    'less than once a month': 'rare',
    'once a week': 'weekly',
    'once a month': 'monthly',
    'multiple times a week': 'very_frequent'
})
```

```python
df['Browsing_Frequency'] = df['Browsing_Frequency'].replace({
    'rarely': 'rare',
    'few times a month': 'low',
    'few times a week': 'medium',
    'multiple times a day': 'high'
})
```

```python
#13.Clean Service_Appreciation and  Improvement_Areas column
df['Service_Appreciation'] = df['Service_Appreciation'].replace({
    '.': 'unknown',
    'all the above': 'multiple_services'
})

df['Improvement_Areas'] = df['Improvement_Areas'].replace({
    '.': 'none',
    'nil': 'none',
    'nothing': 'none',
    'ui': 'app_interface'
})
```

```python
#14.Frequent_Buyer feature
df['Frequent_Buyer'] = (
    df['Purchase_Frequency']
    .str.strip()
    .str.lower()
    .isin(['weekly', 'very frequent'])
    .astype(int)
)

```

```python
#Review_Engagement feature
df['Review_Engagement'] = df['Review_Left'].map({
    'yes': 'engaged',
    'no': 'not_engaged'
})
```

```python
#Shopping_Satisfaction level feature
df['Shopping_Satisfaction_Level'] = pd.cut(
    df['Shopping_Satisfaction'],
    bins=[0, 2, 3, 4, 5],
    labels=['low', 'medium', 'high', 'very_high']
)
```

```python
cols = list(df.columns)

def move_after(col_to_move, target_col):
    cols.insert(
        cols.index(target_col) + 1,
        cols.pop(cols.index(col_to_move))
    )

move_after('Frequent_Buyer', 'Purchase_Frequency')
move_after('Review_Engagement', 'Review_Left')
move_after('Shopping_Satisfaction_Level', 'Shopping_Satisfaction')

df = df[cols]
```
